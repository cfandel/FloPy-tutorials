    <br>GHB - General Head Boundary
    <br>WEL - Well

Description:
Unconfined & transient,with time varying boundaries & general head boundaries

Initial conditions:
<br>Head = 10.0 m across entire model domain
<br>Period 1 (1 day) – steady state with left and right GHB stage = 10.
<br>Period 2 (100 days) – left GHB with stage = 10., right GHB with stage set to 0.
<br>Period 3 (100 days) – pumping well at model center with rate = -100., left and right GHB = 10., and 0.

Import modules

```python
import flopy
import numpy as np
import matplotlib.pyplot as plt
```

Define model properties (domain, grid)

```python
#Model domain & grid parameters:
Lx = 1000.
Ly = 1000.
ztop = 10.
zbot = -50.
nlay = 1
nrow = 10
ncol = 10
delr = Lx / ncol
delc = Ly / nrow
delv = (ztop - zbot) / nlay
botm = np.linspace(ztop, zbot, nlay + 1)
hk = 1.
vka = 1.
sy = 0.1
ss = 1.e-4
laytyp = 1

#Time & stress period parameters:
nper = 3                   #define 3 stress periods
perlen = [1, 100, 100]     #duration of each stress period 
nstp = [1, 100, 100]       #number of timesteps in each stress period
steady = [True, False, False]  #is each stress period steady-state or transient

#Initial head distribution (for BAS package):
# Note that changes from the previous tutorial!
ibound = np.ones((nlay, nrow, ncol), dtype=np.int32)   #creates an integer array filled with ones the size of the model grid (dimensions: z,y,x)
#if ibound < 0, constant head, if = 0, inactive, if > 0 active.
strt = 10. * np.ones((nlay, nrow, ncol), dtype=np.float32)   #creates a float array filled with ones & multiplies by 10
#strt represents starting heads
```

Create time-invariant model objects
<br> (this is a condensed version of Tutorial 1)

```python
# Flopy objects
modelname = 'tutorial2'
mf = flopy.modflow.Modflow(modelname, exe_name='mf2005')
dis = flopy.modflow.ModflowDis(mf, nlay, nrow, ncol, delr=delr, delc=delc,
                               top=ztop, botm=botm[1:],
                               nper=nper, perlen=perlen, nstp=nstp, steady=steady)
bas = flopy.modflow.ModflowBas(mf, ibound=ibound, strt=strt)
lpf = flopy.modflow.ModflowLpf(mf, hk=hk, vka=vka, sy=sy, ss=ss, laytyp=laytyp)
pcg = flopy.modflow.ModflowPcg(mf)
```

GHB (General Head Boundary): Create GHB object
<br> - simulates head-dependent flux boundaries
<br> - boundary conditions are stored in a dictionary keyed to stress periods (starts at 0)
<br> - boundary conditions are in format:  [layer, row, column, stage, conductance]

```python
# Make list for stress period 1
stageleft = 10.       #constant head boundary to be assigned to the left
stageright = 10.
bound_sp1 = []        #creates empty array to fill

#iterate over layers
#note: xrange needs to be changed to range from the tutorial text
for il in range(nlay): 
    #set conductance: C = K*A/dL (combines K, A, and dL terms of Darcy's Law) so K*wetted thickness*columnwidth
    #where is dL?
    condleft = hk * (stageleft - zbot) * delc             
    condright = hk * (stageright - zbot) * delc
    
    #iterate over rows
    for ir in range(nrow):
        bound_sp1.append([il, ir, 0, stageleft, condleft])   #append array of [lay row col head cond] for first column
        bound_sp1.append([il, ir, ncol - 1, stageright, condright])   #same for last column
print('Adding ', len(bound_sp1), 'GHBs for stress period 1.')
#this should be 2x the number of rows (because each row has a left and a right boundary)

# Make list for stress period 2
stageleft = 10.
stageright = 0.
condleft = hk * (stageleft - zbot) * delc
condright = hk * (stageright - zbot) * delc
bound_sp2 = []
for il in range(nlay):
    for ir in range(nrow):
        bound_sp2.append([il, ir, 0, stageleft, condleft])
        bound_sp2.append([il, ir, ncol - 1, stageright, condright])
print('Adding', len(bound_sp2), 'GHBs for stress period 2.')

# We do not need to add a dictionary entry for stress period 3.
# Flopy will automatically take the list from stress period 2 and apply it
# to the end of the simulation, if necessary
stress_period_data = {0: bound_sp1, 1: bound_sp2}

# Create the flopy ghb object
ghb = flopy.modflow.ModflowGhb(mf, stress_period_data=stress_period_data)
```

WEL (Well): Create object for transient WEL package

```python
# Create the well package
# Remember to use zero-based layer, row, column indices!
pumping_rate = -100.     #negative means outflow
#defie well arrays for each stress period, in format: [lay row col Q]
#this puts the well in the center of the model grid, and it doesn't start pumping until the 3rd stress period.
wel_sp1 = [[0, nrow/2 - 1, ncol/2 - 1, 0.]]      
wel_sp2 = [[0, nrow/2 - 1, ncol/2 - 1, 0.]]
wel_sp3 = [[0, nrow/2 - 1, ncol/2 - 1, pumping_rate]]
stress_period_data = {0: wel_sp1, 1: wel_sp2, 2: wel_sp3}
wel = flopy.modflow.ModflowWel(mf, stress_period_data=stress_period_data)
```

OC (Output Control): Create object for OC package

```python
# Output control
spd = {(0, 0): ['print head', 'save head', 'print budget', 'save budget']}
oc = flopy.modflow.ModflowOc(mf, stress_period_data=spd)
```

**Run MODFLOW**

```python
# Write the model input files
mf.write_input()

# Run the model
success, mfoutput = mf.run_model(silent=True, pause=False, report=True)
if not success:
    raise Exception('MODFLOW did not terminate normally.')
```

**Post-Process results**

```python
# Import modules
import matplotlib.pyplot as plt
import flopy.utils.binaryfile as bf

# Create the headfile and budget file objects
# Don't change this
headobj = bf.HeadFile(modelname+'.hds')
times = headobj.get_times()
#cbb = bf.CellBudgetFile(modelname+'.cbc')   #this line doesn't run? identical to tutorial1...
```

Plot using modelmap

```python
# elements using cell budget have been commented out to allow plotting
# Setup contour parameters
levels = np.linspace(0, 10, 11)
extent = (delr/2., Lx - delr/2., delc/2., Ly - delc/2.)
print('Levels: ', levels)
print('Extent: ', extent)

# Calculate (x,y) for well location
wpt = ((float(ncol/2)-0.5)*delr, (float(nrow/2-1)+0.5)*delc)
wpt = (450., 550.)

# Print model stats:
mytimes = [1.0, 101.0, 201.0]
for iplot, time in enumerate(mytimes):
    print('*****Processing time: ', time)
    head = headobj.get_data(totim=time)
    #Print statistics
    print('Head statistics')
    print('  min: ', head.min())
    print('  max: ', head.max())
    print('  std: ', head.std())

    # Extract flow right face and flow front face
    #frf = cbb.get_data(text='FLOW RIGHT FACE', totim=time)[0]
    #fff = cbb.get_data(text='FLOW FRONT FACE', totim=time)[0]

    #Create the plot
    #plt.subplot(1, len(mytimes), iplot + 1, aspect='equal')
    plt.subplot(1, 1, 1, aspect='equal')
    plt.title('stress period ' + str(iplot + 1))


    modelmap = flopy.plot.ModelMap(model=mf, layer=0)
    qm = modelmap.plot_ibound()
    lc = modelmap.plot_grid()
    qm = modelmap.plot_bc('GHB', alpha=0.5)
    cs = modelmap.contour_array(head, levels=levels)
    plt.clabel(cs, inline=1, fontsize=10, fmt='%1.1f', zorder=11)
    #quiver = modelmap.plot_discharge(frf, fff, head=head)


    mfc = 'None'
    if (iplot+1) == len(mytimes):
        mfc='black'
    plt.plot(wpt[0], wpt[1], lw=0, marker='o', markersize=8,
             markeredgewidth=0.5,
             markeredgecolor='black', markerfacecolor=mfc, zorder=9)
    plt.text(wpt[0]+25, wpt[1]-25, 'well', size=12, zorder=12)
    #plt.savefig('tutorial2-{}.png'.format(iplot))                   #uncomment to save as a png file
```
